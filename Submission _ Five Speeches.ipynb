```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
dir_train = "data/train/"
dir_dev = "data/dev/"

VIX_tr = pd.read_json(dir_train + "VIX_1w.json")
EURUSD_tr = pd.read_json(dir_train + "EURUSDV1M_1w.json")

VIX_dev = pd.read_json(dir_dev + "VIX_1w.json")
EURUSD_dev = pd.read_json(dir_dev + "EURUSDV1M_1w.json")
```

# Adding features

```python
# Build one column for each stock

def separate_stocks(X):
    df = X.copy()
    for i in range(1, 21):
        df[f'stock_{i}'] = [stock[i-1] for stock in df['stock']]
    df = df.drop(['stock'], axis=1)
    return df

VIX_tr = separate_stocks(VIX_tr)
EURUSD_tr = separate_stocks(EURUSD_tr)
VIX_dev = separate_stocks(VIX_dev)
EURUSD_dev = separate_stocks(EURUSD_dev)
```

```python
# Listing speeches 

def get_speeches(X):
    df = X.copy()
    rows, cols = df.shape 
    speeches = []
    for i in range(rows):
        dics = df['speech'][i]
        speeches_i = []
        for dic in dics:
            if dic['ECB'] != []:
                speeches_i.append(dic['ECB'][0])
            if dic['FED'] != []:
                speeches_i.append(dic['FED'][0])
        speeches.append(speeches_i)
    df['speech_list'] = speeches
    return df

VIX_tr = get_speeches(VIX_tr)
EURUSD_tr = get_speeches(EURUSD_tr)
VIX_dev = get_speeches(VIX_dev)
EURUSD_dev = get_speeches(EURUSD_dev)
```

```python
## Remove foreign speeches

import spacy
from spacy.language import Language
from spacy_langdetect import LanguageDetector

def get_lang_detector(nlp, name):
    return LanguageDetector()

nlp = spacy.load("en_core_web_sm")
Language.factory("language_detector", func=get_lang_detector)
nlp.add_pipe('language_detector', last=True)

def remove_foreign_lang(X):
    df = X.copy()
    rows, cols = df.shape
    en_speeches = []
    foreign_speeches = []
    for i in range(rows):
        speech_list = df['speech_list'][i]
        for speech in speech_list:
            if speech in foreign_speeches:
                speech_list.remove(speech)
            else:
                sentence = speech[:200]                        # Take only the beginning of the speech
                dic = nlp(speech)._.language
                if dic['language'] != 'en' :
                    speech_list.remove(speech)
                    foreign_speeches.append(speech)
        en_speeches.append(speech_list)
        print(i)
    df['speech_list'] = en_speeches
    return df
```

```python
p = 5

def padding_speeches(X):
    df = X.copy()
    rows, cols = df.shape
    pad_speeches = []
    for i in range(rows):
        speech_list = df['speech_list'][i]
        if speech_list == []:
            speech_list = ['Empty' for i in range(p)]
        else:
            n = len(speech_list)
            if n >= p:
                speech_list = speech_list[n-p : n]
            if n < p:
                for i in range(n, p):
                    speech_list.append(speech_list[n-1])
        pad_speeches.append(speech_list)
    df['speech_list'] = pad_speeches
    return df

VIX_tr = padding_speeches(VIX_tr)
EURUSD_tr = padding_speeches(EURUSD_tr)
VIX_dev = padding_speeches(VIX_dev)
EURUSD_dev = padding_speeches(EURUSD_dev)
```

```python
def negated(word):
    """
    Determine if preceding word is a negation word
    """
    if word.lower() in negate:
        return True
    else:
        return False
```

```python
def tone_count_with_negation_check(dict, article):
    """
    Count positive and negative words with negation check. Account for simple negation only for positive words.
    Simple negation is taken to be observations of one of negate words occurring within three words
    preceding a positive words.
    """
    pos_count = 0
    neg_count = 0
 
    pos_words = []
    neg_words = []
 
    input_words = re.findall(r'\b([a-zA-Z]+n\'t|[a-zA-Z]+\'s|[a-zA-Z]+)\b', article.lower())
 
    word_count = len(input_words)
 
    for i in range(0, word_count):
        if input_words[i] in dict['Negative']:
            neg_count += 1
            neg_words.append(input_words[i])
        if input_words[i] in dict['Positive']:
            if i >= 3:
                if negated(input_words[i - 1]) or negated(input_words[i - 2]) or negated(input_words[i - 3]):
                    neg_count += 1
                    neg_words.append(input_words[i] + ' (with negation)')
                else:
                    pos_count += 1
                    pos_words.append(input_words[i])
            elif i == 2:
                if negated(input_words[i - 1]) or negated(input_words[i - 2]):
                    neg_count += 1
                    neg_words.append(input_words[i] + ' (with negation)')
                else:
                    pos_count += 1
                    pos_words.append(input_words[i])
            elif i == 1:
                if negated(input_words[i - 1]):
                    neg_count += 1
                    neg_words.append(input_words[i] + ' (with negation)')
                else:
                    pos_count += 1
                    pos_words.append(input_words[i])
            elif i == 0:
                pos_count += 1
                pos_words.append(input_words[i])
 
    results = [word_count, pos_count, neg_count, pos_words, neg_words]
 
    return results
```

```python
# Summarization

lemmatizer = WordNetLemmatizer()
important_words = lmdict['Negative'] + lmdict['Positive']

def _rank_sentences(sentences):
    count_results = {}
    for i, sentence in enumerate(sentences):
        input_words = [lemmatizer.lemmatize(t) for t in word_tokenize(sentence)]
        counter = len([value for value in input_words if value in important_words])
        count_results.update({str(i): counter})
        
    rank_dict = OrderedDict(sorted(count_results.items(), key=lambda x: x[1], reverse=True))
    return rank_dict

def _generate_summary(sentences, rank_dict, max_words):
    summary = ''
    keep_sentence = []
    num_words = 0
    sentence_num_words = [len(word_tokenize(sent)) for sent in sentences]
    added_key = []
    
    # Retrieve indices of sentences till it exceeds max number of words
    for key in rank_dict.keys():
        keep_sentence.append(int(key))
        num_words += sentence_num_words[int(key)]
        if num_words > max_words:
            break
    # print('rank_dict: ', rank_dict)
    # print('keep_sentence: ', keep_sentence)
    
    # Once retrieved, add the corresponding sentences in original order
    for key in sorted(keep_sentence):
        summary += ' ' + sentences[key]
        added_key.append(key)
    # print('added_key: ', added_key)
    # print('word_length: ', len(word_tokenize(summary)))
    # print('+------------------------------------------------------+')
    return summary
```

```python
## Function that adds a column to the dataframe with a summary of the last speech

NUM_MAX_WORDS = 100

def summarize(X):
    df = X.copy()
    new_text = []
    for speech_list in df['speech_list']:
        summary_list = []
        for speech in speech_list:
            summary = ''
            # Create a sentence list
            sentences = sent_tokenize(speech)
            # Create a ordered dict by the rank
            rank_dict = _rank_sentences(sentences)
            # Generate summary text by taking higher sentences up to NUM_MAX_WORDS
            summary = _generate_summary(sentences, rank_dict, NUM_MAX_WORDS)
            # Add to the output list
            summary_list.append(summary)
        new_text.append(summary_list)
    df['summary_list'] = new_text
    return df


start_time = time.time()

VIX_tr = summarize(VIX_tr)
EURUSD_tr = summarize(EURUSD_tr)
VIX_dev = summarize(VIX_dev)
EURUSD_dev = summarize(EURUSD_dev)

print("--- %s seconds ---" % (time.time() - start_time))
```

# Embedding summaries

```python
import collections
import string
from tqdm import tqdm
tqdm.pandas(desc="progress-bar")
from nltk.corpus import stopwords
```

```python
# Store the model we want to use

model_name = "mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"

from transformers import RobertaTokenizer, RobertaModel

tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = RobertaModel.from_pretrained(model_name)
```

```python
# tokenize the sentences = divide them into wordpieces.

def tokenize_summaries(X):
    df = X.copy()
    summary_list = df['summary_list']
    tokenized_list = []
    for l_summary in summary_list:
        l_tokenized = []
        for summary in l_summary:
            l_tokenized.append( tokenizer.encode(summary, max_length=100,
                                         truncation=True, add_special_tokens=True, padding='max_length'))
        tokenized_list.append(l_tokenized)
    df['summary_list'] = tokenized_list
    return df

VIX_tr = tokenize_summaries(VIX_tr)
EURUSD_tr = tokenize_summaries(EURUSD_tr)
VIX_dev = tokenize_summaries(VIX_dev)
EURUSD_dev = tokenize_summaries(EURUSD_dev)
```

```python
def separate_summaries(X):
    df = X.copy()
    rows, cols = df.shape
    p = len(df['summary_list'][0])
    for k in range(p):
        summaries_k = []
        for i in range(rows):
            summaries_k.append(df['summary_list'].iloc[i][k])
        df[f'summary_{k+1}'] = summaries_k
    df = df.drop(['summary_list'], axis=1)
    return df

VIX_tr = separate_summaries(VIX_tr)
EURUSD_tr = separate_summaries(EURUSD_tr)
VIX_dev = separate_summaries(VIX_dev)
EURUSD_dev = separate_summaries(EURUSD_dev)
```

```python
# pad all sequences to the same size, to feed the model with a matrix instead of a list of lists (much faster).

def pad_summaries(X):
    df  = X.copy()
    for k in range(1, p+1):
        
        max_len = 0
        tokenized_values = df[f'summary_{k}']
        for summary in tokenized_values:
            if len(summary) > max_len:
                max_len = len(summary)
        padded_k = [summary + [0]*(max_len-len(summary)) for summary in tokenized_values]    
        df[f'padded_{k}'] = padded_k

        # mask to ignore the padded section of the sequences
        df[f'attention_mask_{k}'] = np.where(np.array(padded_k) != 0, 1, 0).tolist()
        
        df= df.drop([f'summary_{k}'], axis=1)
        
    return df

VIX_tr = pad_summaries(VIX_tr)
EURUSD_tr = pad_summaries(EURUSD_tr)
VIX_dev = pad_summaries(VIX_dev)
EURUSD_dev = pad_summaries(EURUSD_dev)
```

```python
# Feed all sequences to the model. Return the weights of the last hidden layer (but we could use any combination of layers)

import torch

import warnings
warnings.filterwarnings('ignore')

def embed_summaries(X):
    df = X.copy()
    
    for k in range (1, p+1):
        
        input_ids = torch.tensor(df[f'padded_{k}'])  
        attention_mask = torch.tensor(df[f'attention_mask_{k}'])
        last_hidden_states = []
        with torch.no_grad():
            for seq, mask in zip(tqdm(input_ids), attention_mask):
                machin = model(seq[None,:], attention_mask=mask[None,:])
                tensor = model(seq[None,:], attention_mask=mask[None,:]).last_hidden_state[0,0]
                last_hidden_states.append(np.mean(tensor.numpy()))
        df[f'lhs_{k}'] = last_hidden_states
        df = df.drop([f'padded_{k}', f'attention_mask_{k}'], axis=1)
    return df

VIX_tr = embed_summaries(VIX_tr)
EURUSD_tr = embed_summaries(EURUSD_tr)
VIX_dev = embed_summaries(VIX_dev)
EURUSD_dev = embed_summaries(EURUSD_dev)
```

```python
def diff_20_19_18(X):
    df = X.copy()
    diff_20 = []
    diff_19 = []
    diff_18 = []
    row, cols = df.shape
    for i in range(row):
        if df['stock_20'][i] > df['stock_19'][i]:
            diff_20.append(1)
        else:
            diff_20.append(0)
        if df['stock_19'][i]> df['stock_18'][i]:
            diff_19.append(1)
        else:
            diff_19.append(0)    
        if df['stock_18'][i]> df['stock_17'][i]:
            diff_18.append(1)
        else:
            diff_18.append(0)
    df['diff_20'] = diff_20
    df['diff_19'] = diff_19
    df['diff_18'] = diff_18
    return df

VIX_tr = diff_20_19_18(VIX_tr)
EURUSD_tr = diff_20_19_18(EURUSD_tr)
VIX_dev = diff_20_19_18(VIX_dev)
EURUSD_dev = diff_20_19_18(EURUSD_dev)
```
