```python
## Function that creates a new column in the dataframe containing the last speech
def get_last_speech(X):
    df = X.copy()
    rows, cols = df.shape
    last_speeches = []
    for i in range(rows):
        dic = df['speech'][i][19]
        if dic['ECB'] != []:
            last_speeches.append(dic['ECB'][0])
        else:
            last_speeches.append(dic['FED'][0])
    df['last speech'] = last_speeches
    return df
```

```python
#import re
#print(re.sub(r"\[\d\].", "", test))
def remove_footnote(speech):
    # find index of second occurence of substring in string
    idx = speech.find("[1]", speech.find("[1]") + 1)
    #idx = speech.find("Footnotes", speech.find("[1]") + 1)
    if idx != -1:
        return speech[:idx]
```
